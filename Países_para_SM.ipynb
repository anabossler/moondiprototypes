```python
def Mahalanobis (data,vector,cov_inv) :
  #data es el dataframe
  #vector es el vector del cual se tomaára la distancia de Mahalanobis
  #cov_inv es la matriz inversa de la matriz de covarianzas
  indices = data.index.values.tolist() #Lista de indices
  A= []
  for i in indices :
    Listapais = data.loc[i, :].values.tolist()
    #print(Listapais)
    vectorpais=np.matrix(Listapais[1:])
    vectorillo = vectorpais-vector
    mult1 = np.matmul(cov_inv,vectorillo.T)
    mult2 = np.matmul(vectorillo,mult1)
    A.append(mult2[0,0])
  return (A)
```

```python
# mat_Ru es el vector de reino unido
# inv_V es la matriz inversa de la matriz de covarianzas

Mahalanobis(bn_data1,mat_RU,inv_V)
```

```python
colim = Mahalanobis(bn_data1,mat_RU,inv_V)
print("Given list is : \n",colim)
# using sorted() and index()
resi = [sorted(colim,reverse=False).index(i) for i in colim]
# printing result
print("list with relative ordering of elements : \n",resi)
```
